### Q2. ê°œë… êµ¬í˜„

#### 2-1. **[ë‚´ì (Dot Product)ê³¼ Norm]**   
##### ë‘ ë²¡í„°ì˜ ìœ ì‚¬ë„ë¥¼ íŒë‹¨í•˜ê±°ë‚˜ ê¸¸ì´ë¥¼ êµ¬í•  ë•Œ ì‚¬ìš©ë˜ëŠ” í•¨ìˆ˜ì…ë‹ˆë‹¤. ë¹ˆì¹¸ì„ ì±„ì›Œ í•¨ìˆ˜ë¥¼ ì™„ì„±í•´ì£¼ì„¸ìš”.

```python
import numpy as np

def calculate_similarity_components(v1, v2):
    """
    v1, v2: numpy array (1D vectors)
    ë°˜í™˜ê°’: ë‘ ë²¡í„°ì˜ ë‚´ì  ê°’, v1ì˜ norm, v2ì˜ norm
    """
    # 1. ë‘ ë²¡í„°ì˜ ë‚´ì (Dot Product)ì„ ê³„ì‚°í•˜ì„¸ìš”.
    dot_prod = np.dot(v1, v2)

    # 2. ê° ë²¡í„°ì˜ í¬ê¸°(Norm, L2 Norm)ë¥¼ ê³„ì‚°í•˜ì„¸ìš”.
    norm_v1 = np.linalg.norm(v1)
    norm_v2 = np.linalg.norm(v2)

    return dot_prod, norm_v1, norm_v2
```

```python
# í•¨ìˆ˜ í™•ì¸
v1 = np.array([3.0, 4.0])  # ì˜ˆì‹œ ë²¡í„° 1
v2 = np.array([12.0, 5.0])  # ì˜ˆì‹œ ë²¡í„° 2
out = calculate_similarity_components(v1, v2)

print(f"ë‚´ì ê°’ : {out[0]}, v1ì˜ norm : {out[1]}, v2ì˜ norm : {out[2]}")
```

#### 2-2. **[ì„ í˜•ë³€í™˜ (í–‰ë ¬ ê³±)]**
##### ì…ë ¥ ë²¡í„°ë¥¼ ë‹¤ë¥¸ ì°¨ì›ì˜ ê³µê°„ìœ¼ë¡œ ë§¤í•‘í•˜ëŠ” ì„ í˜•ë³€í™˜ ê³¼ì •ì…ë‹ˆë‹¤. ë¹ˆì¹¸ì„ ì±„ì›Œ í•¨ìˆ˜ë¥¼ ì™„ì„±í•´ì£¼ì„¸ìš”.

```python
from os import waitpid
def linear_transformation(x, W, b):
    """
    x: ì…ë ¥ ë²¡í„° (input features)
    W: ê°€ì¤‘ì¹˜ í–‰ë ¬ (weight matrix)
    b: í¸í–¥ ë²¡í„° (bias)
    ë°˜í™˜ê°’: ì„ í˜•ë³€í™˜ ê²°ê³¼ z = Wx + b
    """
    # 1. í–‰ë ¬ ê³±(Matrix Multiplication) Wxë¥¼ ìˆ˜í–‰í•˜ì„¸ìš”.
    # íŒíŠ¸: numpyì˜ í–‰ë ¬ ê³± ì—°ì‚°ì(@) ë˜ëŠ” np.dot, np.matmul ì‚¬ìš©
    wx = W@x

    # 2. í¸í–¥(bias)ì„ ë”í•˜ì„¸ìš”.
    z = wx+b

    return z
```

```python
# í•¨ìˆ˜ í™•ì¸
x = np.array([1.0, 2.0])  # ì˜ˆì‹œ input vector
W = np.array([[0.5, 1.0], [1.5, -0.5]])  # ì˜ˆì‹œ weight matrix
b = np.array([0.1, -0.2])  # ì˜ˆì‹œ bias vector

print(linear_transformation(x, W, b))
```

#### 2-3. **[ë¹„ì„ í˜•ì„± (ReLU í•¨ìˆ˜)]**

```python
def relu_function(z):
    """
    z: ì„ í˜•ë³€í™˜ì„ ê±°ì¹œ ê°’
    ë°˜í™˜ê°’: 0ë³´ë‹¤ ì‘ìœ¼ë©´ 0, 0ë³´ë‹¤ í¬ë©´ ê·¸ëŒ€ë¡œ ë°˜í™˜ (max(0, z))
    """
    # numpyì˜ maximum í•¨ìˆ˜ ë“±ì„ ì‚¬ìš©í•˜ì—¬ ReLUë¥¼ êµ¬í˜„í•˜ì„¸ìš”.
    a = np.maximum(0, z)
    return a
```

### Q3. ì‘ìš© ë¬¸ì œ

ìœ„ 2ë²ˆ ë¬¸ì œì—ì„œ ì‘ì„±í•œ í˜ìˆ˜ë“¤(calculate_similarity_components, linear_transformation, relu_function)ì„ í™œìš©í•˜ì—¬ ì•„ë˜ ë¬¸ì œë¥¼ í•´ê²°í•˜ëŠ” ì½”ë“œë¥¼ ì‘ì„±í•˜ì„¸ìš”.

#### ê°„ë‹¨í•œ ìœ ì‚¬ë„ ê¸°ë°˜ ë¶„ë¥˜ê¸° êµ¬í˜„.
ì…ë ¥ ë°ì´í„° xê°€ ì£¼ì–´ì¡Œì„ ë•Œ, 1ì°¨ì ìœ¼ë¡œ ì„ í˜• ë³€í™˜ê³¼ ë¹„ì„ í˜• ë³€í™˜ì„ ê±°ì³ íŠ¹ì§•(feature)ì„ ì¶”ì¶œí•˜ê³ , ì´ ì¶”ì¶œëœ íŠ¹ì§•ì´ ê¸°ì¤€ ë²¡í„°(reference vector)ì™€ ì–¼ë§ˆë‚˜ ìœ ì‚¬í•œì§€ ì½”ì‚¬ì¸ ìœ ì‚¬ë„(Cosine Similarity)ë¡œ íŒë‹¨í•©ë‹ˆë‹¤.

```python
def forward_and_evaluate(x, W, b, ref_vector):

    # 1. ì„ í˜• ë³€í™˜
    wx = W@x
    z = wx+b

    # 2. ë¹„ì„ í˜• í™œì„±í™”(ReLU)
    a = np.maximum(0, z)

    # 3. ì½”ì‚¬ì¸ ìœ ì‚¬ë„ ê³„ì‚°
    dot_prod = np.dot(a, ref_vector)
    norm_a = np.linalg.norm(a)
    norm_ref = np.linalg.norm(ref_vector)
    cos_sim = dot_prod / (norm_a * norm_ref)

    # ì½”ì‚¬ì¸ ìœ ì‚¬ë„ ê³µì‹ ì ìš© (ë¶„ëª¨ê°€ 0ì´ ì•„ë‹˜ì„ ê°€ì •)

    return cos_sim
```

```python
# í…ŒìŠ¤íŠ¸ ì½”ë“œ
x = np.array([1, 2])
W = np.array([[1, -1], [2, 0], [0, 1]]) # 3x2 í–‰ë ¬
b = np.array([0, 1, -1])
ref = np.array([1, 1, 0])
print(forward_and_evaluate(x, W, b, ref))
```
