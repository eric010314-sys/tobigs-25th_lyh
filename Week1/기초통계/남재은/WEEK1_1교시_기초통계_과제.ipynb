## A1. 답안 작성  
답변:
A/B TEST란 웹사이트, 앱 등에서 두 가지 버전의 콘텐츠를 무작위 사용
자 집단에 노출시켜 어떤 버전이 더 나은 성과를 내는지 비교하여 둘 중 더 나은 버전을 통계적인 기반으로 선택하기 위한 실험이다.

A/B TEST 시 초두효과에 유의해야한다. 이는 실제 웹사이트, 앱 등의 환경이 바뀌었을 때 사용자가 기존의 버전(A)에 비해 새로운 변화(B)가 낯설게 느껴져 클릭율 등이 저조하게 나오는 것을 의미한다.

초두효과와 반대되는 효과로 신기함 효과가 있다. 사용자가 오히려 새로운 변화에 흥미를 느끼고 클릭해 발생하는 효과이다.

예를 들어, 한 무신사 어플에서 상품 가격을 빨간색으로 바꾸는 A/B TEST를 진행한다고 치자. 사용자 집단 m에서는 빨간 가격표가 낯설어 앱을 떠나는 비율이 높은 초두효과가 발생하고, 사용자 집단 n에서는 신기함 효과가 발생해 더 많은 클릭이 발생했다. 이런 효과들을 최소화하기 위해 실험 기간을 길게 두어 실험을 진행하여야 한다.

## Q2-2. 실험에 필요한 최소 test users 수를 구해봅시다.
유의수준 0.05, 검정력 80% 하에 수행해주세요.  


아래 공식 참고해서, 최소 test users 수를 구하는 코드 작성해주세요.

추가적으로 n은 산출하는데에 가정한 p1, p2도 같이 작성해주세요.

$$n = \frac{(Z_{\alpha/2} + Z_{\beta})^2 \cdot [p_1(1-p_1) + p_2(1-p_2)]}{(p_1 - p_2)^2}$$

## A2-2. 아래 질문에 대한 답과 코드를 채워주세요.

p1, p2는 각각 무엇을 의미하는 확률이고, 어떤 값으로 가정하셨나요?   
또한 위 공식을 보고 n을 구하기 위한 코드를 작성해주세요. 단, 라이브러리는 z값을 구하는 용도의 것만 사용해주세요.


### 답변:

* $p_1, p_2$의 의미와 가정된 값$p_1$ : 대조군(기존 안)의 전환율이다. 현재 디자인에서의 평균적인 성과 지표를 의미한다.
* $p_2$ : 실험군(Experimental Group, 변경 안)의 목표 전환율이다. 우리가 실험을 통해 기대하는 개선된 성과 지표이다.

**[가정된 값]** 예시 코드를 작성하기 위해 임의로 다음과 같이 설정했다.
* $p_1 = 0.20$ (20%): 기존 공모전 상세페이지의 클릭율이 20%라고 가정.
* $p_2 = 0.22$ (22%): 디자인 변경을 통해 클릭율이 2%p 상승할 것이라고 기대하여 22%로 가정.

```python
# A2-2 코드작성. 나온 n은 아래 Minimum Sample Size에 기재.

import math
from scipy import stats

def calculate_min_sample_size(alpha, power, p1, p2):
    """
    사용자가 제시한 공식을 사용하여 최소 표본 크기(n)를 계산하는 함수
    n = ((Z_alpha/2 + Z_beta)^2 * (p1(1-p1) + p2(1-p2))) / (p1 - p2)^2
    """

    # 1. Z값 구하기 (표준정규분포 역함수 사용)
    # 유의수준 alpha에 대한 임계값 (양측 검정, alpha/2)
    # alpha=0.05일 때 약 1.96
    z_alpha = stats.norm.ppf(1 - alpha / 2)

    # 검정력(1-beta)에 대한 임계값
    # Power=0.8일 때 약 0.84
    z_beta = stats.norm.ppf(power)

    # 2. 공식 대입
    # 분자: (Z_alpha + Z_beta)^2 * 분산항 합
    numerator = (z_alpha + z_beta)**2 * (p1 * (1 - p1) + p2 * (1 - p2))

    # 분모: (두 확률 차이)^2
    denominator = (p1 - p2)**2

    # 3. 최종 n 계산
    n = numerator / denominator

    return n, z_alpha, z_beta

# --- 설정값 입력 ---
alpha = 0.05  # 유의수준 5%
power = 0.80  # 검정력 80%
p1 = 0.20     # 대조군 전환율 가정 (10%)
p2 = 0.22     # 실험군 전환율 가정 (12%)

# 계산 실행
required_n, z_a, z_b = calculate_min_sample_size(alpha, power, p1, p2)

print(f"=== 입력 변수 ===")
print(f"유의수준(alpha): {alpha}")
print(f"검정력(power): {power}")
print(f"기존 전환율(p1): {p1}")
print(f"목표 전환율(p2): {p2}")
print(f"\n=== 중간 계산 값 ===")
print(f"Z_alpha/2: {z_a:.4f}")
print(f"Z_beta:    {z_b:.4f}")
print(f"\n=== 결과 ===")
print(f"그룹 당 최소 필요 표본 수 (n): {required_n:.4f}")
print(f"올림 처리 후 (n): {math.ceil(required_n)} 명")










#########
```

[실험 가설]  
귀무가설 ($H_0$): 메인 홈 검색 모듈의 위치(하단 vs 최상단)와 상세 페이지 진입 여부 사이에는 상관관계가 없다. 즉, 두 그룹의 상세 페이지 진입률($CVR$)은 동일하다.$$H_0: p_A = p_B$$대립가설 ($H_1$): 메인 홈 검색 모듈을 최상단에 노출하는 것은 상세 페이지 진입 여부에 영향을 미친다. 즉, 실험군(B)의 상세 페이지 진입률이 대조군(A)보다 높을 것이다.$$H_1: p_A < p_B$$

## Q3-1. 위 가설에 적합한 검정 이름과 왜 해당 검정을 써야하는지 이유도 작성해주세요. (t test와 카이제곱검정 중)

## A3-1.  검정 이름과 이유를 설명해주세요.

답변: 위 가설에 필요한 검정은 카이제곱 검정  이고,

이유: 분석하려는 결과 데이터인 '클릭 여부(Click / No Click)'가 연속된 숫자가 아닌 **범주형 데이터(Categorical Data)**이기 때문입니다.

우리는 두 그룹의 '평균 점수(Mean)'를 비교하는 것이 아니라, 클릭한 사람의 '빈도(Frequency) 혹은 비율'이 서로 다른지를 확인해야 하므로, t-test가 아닌 카이제곱 검정을 사용해야 합니다.

<br>

```python
'''''''''''''''''''''''''''''''''''''''
제가 세운 A/B TEST 설계 실험 코드입니다.
'''''''''''''''''''''''''''''''''''''''

import pandas as pd
from scipy.stats import chi2_contingency

# 1. 관측 데이터 입력 (실제 실험 결과 수치로 변경하세요)
# [상세진입 성공 수, 상세진입 실패 수]
control_results = [11500, 34500]  # 대조군(A)
variant_results = [12190, 33810]  # 실험군(B)

# 2. 분할표(Contingency Table) 생성
data = [control_results, variant_results]

# 3. 카이제곱 검정 수행
chi2, p_value, dof, expected = chi2_contingency(data)

# 4. 결과 출력
print(f"카이제곱 통계량: {chi2:.4f}")
print(f"P-value: {p_value:.10f}")

alpha = 0.05
if p_value < alpha:
    print(f"결과: P-value가 {alpha}보다 작으므로 귀무가설을 기각합니다.")
    print("=> 검색바 최상단 노출이 상세 페이지 진입률에 유의미한 변화를 주었습니다.")
else:
    print(f"결과: P-value가 {alpha}보다 크므로 귀무가설을 기각할 수 없습니다.")
    print("=> 검색바 위치 변경의 효과가 통계적으로 증명되지 않았습니다.")
```

실제 데이터가 없어서, 저도 필요한 데이터는 가정했답니다.

## Q3-4. 위에서 답한 것을 토대로, 검정을 수행해주세요. (t test와 카이제곱 검정 중)


## A3-4 아래 셀에 코드 작성해주세요. 그리고!!!
그리고 나온 값을 통해 수행하신 a/b test 결과를 설정하신 가설에 맞게 해석해주세요.

```python
### A3-4. 코드작성

import pandas as pd
from scipy.stats import chi2_contingency

# 1. 관측 데이터 입력 (무한 스크롤 실험 결과라고 가정)
# Control(A): 기존 스크롤 방식 [클릭 11500, 미클릭 34500] -> 총 46,000명 (CTR 25.0%)
# Variant(B): 무한 스크롤 방식 [클릭 12190, 미클릭 33810] -> 총 46,000명 (CTR 26.5%)
control_results = [11500, 34500]
variant_results = [12190, 33810]

# 2. 분할표(Contingency Table) 생성
data = [control_results, variant_results]

# 3. 카이제곱 검정 수행
chi2, p_value, dof, expected = chi2_contingency(data)

# 4. 결과 출력
print(f"=== 무한 스크롤 A/B Test 결과 ===")
print(f"A그룹(기존) 클릭률: {control_results[0] / sum(control_results):.2%}")
print(f"B그룹(무한) 클릭률: {variant_results[0] / sum(variant_results):.2%}")
print(f"-----------------------")
print(f"카이제곱 통계량: {chi2:.4f}")
print(f"P-value: {p_value:.10f}")

alpha = 0.05
print(f"-----------------------")
if p_value < alpha:
    print(f"결과: P-value가 {alpha}보다 작으므로 귀무가설을 기각합니다.")
    print("해석: 무한 스크롤 도입이 공모전 클릭률에 통계적으로 유의미한 긍정적 변화를 주었습니다.")
else:
    print(f"결과: P-value가 {alpha}보다 크므로 귀무가설을 기각할 수 없습니다.")
    print("해석: 무한 스크롤 도입에 따른 클릭률 차이가 통계적으로 증명되지 않았습니다.")

#############
```

**결과해석**

답변:통계적으로 유의미한 수준에서 귀무가설을 기각합니다. 즉, 무한 스크롤 형태로의 변경은 사용자의 공모전 탐색 및 클릭을 유도하는 데 효과가 있었다고 판단할 수 있습니다. 따라서 B안을 전체 배포하는 것을 제안합니다
