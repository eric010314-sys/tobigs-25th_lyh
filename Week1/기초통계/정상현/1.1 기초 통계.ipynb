답변: A/B Test란 사용자를 대조군과 실험군의 두 집단으로 나누어 두 집단을 동시에 노출한 뒤 사전에 정의한 지표의    평균 차이가 유의미한지를 평가해 변경의 효과를 추정하는 방법이다.
     A/B Test시 유의할 사항에는 집단 배정의 무작위성 유지, 측정 단위의 일관성, 가설과 주요 지표의 고정, 초두효과 등이 존재한다. 이때 초두효과란 사용자가 새롭거나 변경된 기능을 처음으로 접했을 때의 낯섦 때문에 행동이 일시적으로 변하는 효과이다. 이러한 초반의 데이터만으로는 초두효과와 장기효과를 구분하기 힘들기 때문에 충분한 기간을 두고 데이터를 분석하는 것이 적절하다.

## Q2-2. 실험에 필요한 최소 test users 수를 구해봅시다.
유의수준 0.05, 검정력 80% 하에 수행해주세요.  


아래 공식 참고해서, 최소 test users 수를 구하는 코드 작성해주세요.

추가적으로 n은 산출하는데에 가정한 p1, p2도 같이 작성해주세요.

$$n = \frac{(Z_{\alpha/2} + Z_{\beta})^2 \cdot [p_1(1-p_1) + p_2(1-p_2)]}{(p_1 - p_2)^2}$$

## A2-2. 아래 질문에 대한 답과 코드를 채워주세요.

p1, p2는 각각 무엇을 의미하는 확률이고, 어떤 값으로 가정하셨나요?   
또한 위 공식을 보고 n을 구하기 위한 코드를 작성해주세요. 단, 라이브러리는 z값을 구하는 용도의 것만 사용해주세요.


### 답변:
p1: 기존 추천 정렬의 클릭률 (가정 : 0.1)
p2: 최신성 가중치 적용 시 기대 클릭률 (가정 : 0.13)

```python
# A2-2 코드작성. 나온 n은 아래 Minimum Sample Size에 기재.
from scipy.stats import norm
import math

p1 = 0.10   # 기존 클릭률
p2 = 0.13   # 개선 클릭률

alpha = 0.05
power = 0.80
beta = 1 - power

# Z값
z_alpha = norm.ppf(1 - alpha/2)
z_beta = norm.ppf(1 - beta)

# 공식에 필요한 값 계산
num1 = (z_alpha + z_beta)**2 * (p1*(1-p1) + p2*(1-p2))
num2 = (p1 - p2)**2

n = num1 / num2

print("p1:", p1)
print("p2:", p2)
print("Minimum sample size:", math.ceil(n))
#########
```

[실험 가설]  
귀무가설 ($H_0$): 메인 홈 검색 모듈의 위치(하단 vs 최상단)와 상세 페이지 진입 여부 사이에는 상관관계가 없다. 즉, 두 그룹의 상세 페이지 진입률($CVR$)은 동일하다.$$H_0: p_A = p_B$$대립가설 ($H_1$): 메인 홈 검색 모듈을 최상단에 노출하는 것은 상세 페이지 진입 여부에 영향을 미친다. 즉, 실험군(B)의 상세 페이지 진입률이 대조군(A)보다 높을 것이다.$$H_1: p_A < p_B$$

## A3-1.  검정 이름과 이유를 설명해주세요.

답변: 위 가설에 필요한 검정은 카이제곱 검정 이고,

이유는 검색 모듈 위치와 상세 페이지 진입 여부가 모두 범주형 변수이므로 두 집단의 진입률이 동일한지 여부는 각 그룹의 성공 빈도로 만든 분할표에서 독립성 검정을 하는 카이제곱 검정이 적합하기 때문이다.

<br>

```python
'''''''''''''''''''''''''''''''''''''''
제가 세운 A/B TEST 설계 실험 코드입니다.
'''''''''''''''''''''''''''''''''''''''

import pandas as pd
from scipy.stats import chi2_contingency

# 1. 관측 데이터 입력 (실제 실험 결과 수치로 변경하세요)
# [상세진입 성공 수, 상세진입 실패 수]
control_results = [11500, 34500]  # 대조군(A)
variant_results = [12190, 33810]  # 실험군(B)

# 2. 분할표(Contingency Table) 생성
data = [control_results, variant_results]

# 3. 카이제곱 검정 수행
chi2, p_value, dof, expected = chi2_contingency(data)

# 4. 결과 출력
print(f"카이제곱 통계량: {chi2:.4f}")
print(f"P-value: {p_value:.10f}")

alpha = 0.05
if p_value < alpha:
    print(f"결과: P-value가 {alpha}보다 작으므로 귀무가설을 기각합니다.")
    print("=> 검색바 최상단 노출이 상세 페이지 진입률에 유의미한 변화를 주었습니다.")
else:
    print(f"결과: P-value가 {alpha}보다 크므로 귀무가설을 기각할 수 없습니다.")
    print("=> 검색바 위치 변경의 효과가 통계적으로 증명되지 않았습니다.")
```

실제 데이터가 없어서, 저도 필요한 데이터는 가정했답니다.  

이제 25기 분들의 A/B TEST를 수행해보시죠!

## Q3-4. 위에서 답한 것을 토대로, 검정을 수행해주세요. (t test와 카이제곱 검정 중)


## A3-4 아래 셀에 코드 작성해주세요. 그리고!!!
그리고 나온 값을 통해 수행하신 a/b test 결과를 설정하신 가설에 맞게 해석해주세요.

```python
### A3-4. 코드작성
import pandas as pd
from scipy.stats import chi2_contingency

# 1. 관측 데이터 입력 (실제 실험 결과 수치로 변경하세요)
# [상세진입 성공 수, 상세진입 실패 수]
control_results = [180, 1592]  # 대조군(A)
variant_results = [230, 1542]  # 실험군(B)

# 2. 분할표(Contingency Table) 생성
data = [control_results, variant_results]

# 3. 카이제곱 검정 수행
chi2, p_value, dof, expected = chi2_contingency(data)

# 4. 결과 출력
print(f"카이제곱 통계량: {chi2:.4f}")
print(f"P-value: {p_value:.10f}")

alpha = 0.05
if p_value < alpha:
    print(f"결과: P-value가 {alpha}보다 작으므로 귀무가설을 기각합니다.")
    print("=> 최신성 가중치 부여가 상세 페이지 진입률에 유의미한 변화를 주었습니다.")
else:
    print(f"결과: P-value가 {alpha}보다 크므로 귀무가설을 기각할 수 없습니다.")
    print("=> 최신성 가중치 부여가 효과가 통계적으로 증명되지 않았습니다.")










#############
```

**결과해석**

답변: 최신성 가중치를 반영한 추천 정렬은 기존 정렬 대비 상세보기 클릭률에 통계적으로 유의한 차이를 만든다고 해석할 수 있다.
