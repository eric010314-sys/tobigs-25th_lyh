## A1. 답안 작성  
답변:

A/B TEST 는 전체 사용자를 랜덤으로 2개의 집단으로 나눈 뒤, A안과 B안 중 어떤 것이 KPI를 더 잘 달성할 수 있는지 확인하는 방법이다.

예를 들면 현재 내가 운영중인 서비스의 구독버튼을 빨간색(A안)으로 할지 혹은 초록색(B안)으로 할지를 일정 기간동안 테스팅을 수행하고, 만약 A안이 클릭율이 높은 더 좋은 방법이라면 해당 안을 활용하는 것을 의미한다.

그러나 A/B TEST 수행 시에는 초두효과와 신기효과에 유의해야한다.

**1. 초두효과**


*   유저는 기존 방식에 익숙하기 때문에 새로운 변화에 적응하는데 시간이 걸려 초기에는 부정적인 반응을 얻을 수 있는 것을 의미한다.
*   ex) 구매버튼이 기존에 우측에 있었으나 좌측으로 이동한 경우 사용자가 혼동을 느껴 구매율이 감소할 수 있다
*   대안 : 실험기간을 길게 설정하고 초기에 사용성 저하가 발생할 수 있음을 염두하고 평가를 진행한다.


**2. 신기효과**


*   새로운 기능이 등장하면 유저가 호기심을 가져 일시적으로 높은 관심을 보여 긍정적인 반응을 얻는 경우를 의미한다.
*   대안 : 장기적인 사용자 패턴을 분석해서 평가를 진행한다.

## Q2-2. 실험에 필요한 최소 test users 수를 구해봅시다.
유의수준 0.05, 검정력 80% 하에 수행해주세요.  


아래 공식 참고해서, 최소 test users 수를 구하는 코드 작성해주세요.

추가적으로 n은 산출하는데에 가정한 p1, p2도 같이 작성해주세요.

$$n = \frac{(Z_{\alpha/2} + Z_{\beta})^2 \cdot [p_1(1-p_1) + p_2(1-p_2)]}{(p_1 - p_2)^2}$$

## A2-2. 아래 질문에 대한 답과 코드를 채워주세요.

p1, p2는 각각 무엇을 의미하는 확률이고, 어떤 값으로 가정하셨나요?   
또한 위 공식을 보고 n을 구하기 위한 코드를 작성해주세요. 단, 라이브러리는 z값을 구하는 용도의 것만 사용해주세요.


### 답변:

```python
from numpy import power
import math
from scipy import stats

def calculate_min_test_users(alpha, power, p1, p2):

  """
  alpha : 유의 수준 0.05
  power : 검정력 0.80
  p1 : 대조군(A안)의 예상 전환율
  p2 : 실험군(B안)의 예상 전환율
  """

  z_alpha = stats.norm.ppf(1-alpha / 2)
  z_beta = stats.norm.ppf(power)

  #분자 계산
  a = (z_alpha + z_beta) ** 2 * (p1 * (1-p1) + p2 * (1-p2))

  #분모 계산
  b = (p1 - p2) ** 2

  n = a / b
  return math.ceil(n)

alpha_value = 0.05
power_value = 0.80
p1_value = 0.20
p2_value = 0.25

n = calculate_min_test_users(alpha_value, power_value, p1_value, p2_value)
print(f"그룹당 필요한 최소 표본 수 (n): {n} 명")
print(f"총 필요한 표본 수 (A/B 합계): {n * 2} 명")
```

# Q3. A/B TEST를 수행해봅시다.

1-pager를 작성했으면 직접 test도 수행해봐야겠죠!  
위에 제가 만들었던 1-pager를 바탕으로 검정을 해볼게요.  
대표적인 검정방법에는 t-test가 있고 카이제곱 겁정이 있습니다.  


---



[실험 가설]  
귀무가설 ($H_0$): 메인 홈 검색 모듈의 위치(하단 vs 최상단)와 상세 페이지 진입 여부 사이에는 상관관계가 없다. 즉, 두 그룹의 상세 페이지 진입률($CVR$)은 동일하다.$$H_0: p_A = p_B$$대립가설 ($H_1$): 메인 홈 검색 모듈을 최상단에 노출하는 것은 상세 페이지 진입 여부에 영향을 미친다. 즉, 실험군(B)의 상세 페이지 진입률이 대조군(A)보다 높을 것이다.$$H_1: p_A < p_B$$


<br>

## A3-1.  검정 이름과 이유를 설명해주세요.

답변: 위 가설에 필요한 검정은 카이제곱 검정 이다.  

상세페이지의 진입 여부는 진입함, 진입하지 않음 2개의 선택지를 가지는 범주형 데이터이다.

카이제곱 검정은 범주형데이터를, t-test 는 연속형 데이터를 검증하는데 적합한 방법이다.

따라서 위의 가설에 적합한 검정 방법은 두 그룹간 진입률이 다른지를 확인하는 방법이어야 한다. 이 경우 카이제곱 검정이 적합하다.

```python
'''''''''''''''''''''''''''''''''''''''
제가 세운 A/B TEST 설계 실험 코드입니다.
'''''''''''''''''''''''''''''''''''''''

import pandas as pd
from scipy.stats import chi2_contingency

# 1. 관측 데이터 입력 (실제 실험 결과 수치로 변경하세요)
# [상세진입 성공 수, 상세진입 실패 수]
control_results = [11500, 34500]  # 대조군(A)
variant_results = [12190, 33810]  # 실험군(B)

# 2. 분할표(Contingency Table) 생성
data = [control_results, variant_results]

# 3. 카이제곱 검정 수행
chi2, p_value, dof, expected = chi2_contingency(data)

# 4. 결과 출력
print(f"카이제곱 통계량: {chi2:.4f}")
print(f"P-value: {p_value:.10f}")

alpha = 0.05
if p_value < alpha:
    print(f"결과: P-value가 {alpha}보다 작으므로 귀무가설을 기각합니다.")
    print("=> 검색바 최상단 노출이 상세 페이지 진입률에 유의미한 변화를 주었습니다.")
else:
    print(f"결과: P-value가 {alpha}보다 크므로 귀무가설을 기각할 수 없습니다.")
    print("=> 검색바 위치 변경의 효과가 통계적으로 증명되지 않았습니다.")
```

실제 데이터가 없어서, 저도 필요한 데이터는 가정했답니다.  

이제 25기 분들의 A/B TEST를 수행해보시죠!

위에서 가정한 내용에 따르면 두 그룹 간의 CTR 비율 차이를 검증하는 것을 목표로 했으므로, z-test 를 활용하는 것이 적합하다.

## Q3-4. 위에서 답한 것을 토대로, 검정을 수행해주세요. (t test와 카이제곱 검정 중)


## A3-4 아래 셀에 코드 작성해주세요. 그리고!!!
그리고 나온 값을 통해 수행하신 a/b test 결과를 설정하신 가설에 맞게 해석해주세요.

```python
from statsmodels.stats.proportion import proportions_ztest
import numpy as np

""" 데이터 """
# cnt : 성공 횟수 (클릭 수)
# nobs : 전체 시도 횟수 (전체 노출 수 )

#A그룹 (기존 방식) : 20%
#B그룹 (변경 방식) : 25%

count = np.array([590, 526])
nobs = np.array([2182, 2182])

# z-test 수행
z_stat, p_value = proportions_ztest(count=count, nobs=nobs, alternative = 'larger')

print(f"Z-statistic: {z_stat:.4f}")
print(f"p-value: {p_value:.4f}")

print("-" * 30)
alpha = 0.05
if p_value < alpha:
    print(f"결과: P-value가 {alpha}보다 작으므로 귀무가설을 기각합니다.")
    print("=> 홈페이지 버튼 추가가 CTR에 유의미한 변화를 주었습니다.")
else:
    print(f"결과: P-value가 {alpha}보다 크므로 귀무가설을 기각할 수 없습니다.")
    print("=> 홈페이지 버튼 추가 효과가 통계적으로 증명되지 않았습니다.")
```

**결과해석**

답변:

z-test 검정을 수행한 결과 P-value 값이 1.32% 수준으로 확인되었다. 이는 CTR 차이가 우연이었을 확률이 해당 값임을 나타낸다. 또한 유의수준 5%에 비해 크게 작은 수치로 효과를 입증할 수 있다.

z-statistic 수치의 경우 약 2.22로 A안과 B안의 데이터가 겹치지 않고 명확히 분리되었음을 확인할 수 있다.

이런 결과를 볼 때 귀무가설을 기각할 수 있다고 판단되어진다.
